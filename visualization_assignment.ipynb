 # Importaciones en la parte superior (como requiere el ejercicio)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Configuración de estilo


```python
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

#1. Gráfico de Líneas (Series Temporales)
###Datos generados - tendencia de temperatura mensual

```python
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
temperaturas = [22, 24, 26, 28, 30, 32]
precipitacion = [80, 70, 60, 40, 20, 10]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(meses, temperaturas, marker='o', linestyle='--', color='r', label='Temperatura (°C)')
ax.set_xlabel('Mes')
ax.set_ylabel('Temperatura (°C)')
ax.set_title('Tendencia de Temperatura Mensual')
ax.legend()
ax.grid(True)
```

# Guardar imagen


```python
plt.savefig('imgs/line_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

**Explicación**: Este gráfico de líneas muestra la tendencia de temperatura a lo largo de los meses.
Elegí esta visualización porque es ideal para mostrar cambios a lo largo del tiempo.
La línea roja con marcadores hace que sea fácil seguir la tendencia ascendente.

# 2. Gráfico de Barras (Comparación)
### Datos de ventas por producto

```python
productos = ['Laptops', 'Tablets', 'Teléfonos', 'Accesorios']
ventas = [120, 85, 150, 65]

plt.figure(figsize=(10, 6))
bars = plt.bar(productos, ventas, color=['#4C72B0', '#55A868', '#C44E52', '#8172B2'])
plt.xlabel('Productos')
plt.ylabel('Ventas (unidades)')
plt.title('Ventas por Producto en el Último Trimestre')
plt.xticks(rotation=45)

# Añadir valores encima de las barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')
```

#Guardar imagen


```python
plt.savefig('bar_chart.png', dpi=300, bbox_inches='tight')
plt.show()
```

**Explicación**: Este gráfico de barras compara las ventas entre diferentes categorías de productos.
Las barras coloreadas y los valores numéricos hacen que las comparaciones sean inmediatamente evidentes.
Está bueno para mostrar diferencias entre categorías discretas.

#3. Histograma (Distribución)
##Distribución de edades

```python
np.random.seed(42)
edades = np.random.normal(35, 10, 1000)

plt.figure(figsize=(10, 6))
plt.hist(edades, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de Edades en una Población')
plt.grid(axis='y', alpha=0.75)
```

#Guardar imagen

```python
plt.savefig('imgs/histogram.png', dpi=300, bbox_inches='tight')
plt.show()
```

**Explicación**: Este histograma muestra la distribución de edades en una población ficticia.
Elegí 20 bins para balancear detalle y legibilidad.
La distribución normal generada muestra claramente la concentración alrededor de los 35 años.


#4. Diagrama de Dispersión
##Datos de relación entre horas de estudio y calificaciones

```python
horas_estudio = [2, 3, 5, 7, 8, 10, 12, 15, 16, 18]
calificaciones = [50, 55, 70, 75, 80, 85, 88, 90, 92, 95]

plt.figure(figsize=(10, 6))
plt.scatter(horas_estudio, calificaciones,
            c=calificaciones, cmap='viridis',
            s=100, alpha=0.7, edgecolors='w', linewidth=1)

plt.xlabel('Horas de Estudio por Semana')
plt.ylabel('Calificación Final (%)')
plt.title('Relación entre Horas de Estudio y Calificaciones')
plt.colorbar(label='Calificación (%)')
```

#Guardar imagen

```python
plt.savefig('imgs/scatter_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

**Explicación**: Este diagrama de dispersión muestra la relación positiva entre horas de estudio y calificaciones.
Los puntos coloreados por calificación y el gradiente de color ayudan a visualizar la correlación.
Es ideal para mostrar relaciones entre variables continuas.

#5. Gráfico Personalizado
##Para multiples datos (me ayudo Chat GPT)

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

# Primer subplot
ax1.plot(x, y1, color='blue', label='sen(x)')
ax1.set_title('Funciones Trigonométricas')
ax1.legend()

# Segundo subplot
ax2.plot(x, y2, color='red', label='cos(x)')
ax2.legend()

# Tercer subplot
ax3.plot(x, y3, color='green', label='tan(x)')
ax3.legend()
ax3.set_xlabel('Valores de x')
```

#Guardar imagen


```python
plt.tight_layout()
plt.savefig('imgs/custom_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

**Explicación**: Este gráfico personalizado muestra tres funciones trigonométricas en subplots apilados.
La disposición vertical facilita la comparación entre las curvas.
Elegí esta visualización porque en analisis hay de estas funciones.
